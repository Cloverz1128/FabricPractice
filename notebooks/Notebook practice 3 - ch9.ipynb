```python
dfActual = spark.read.format("csv").option("header","true").load("Files/MtoMActual.csv")
dfActual = dfActual.select(dfActual.Country, dfActual.Actual.cast('int')) \
                    .groupBy("Country").sum("Actual").withColumnRenamed("sum(Actual)", "ActualTotal")
dfActual.write.mode("overwrite").format("delta").saveAsTable("MtoMActualSum")
display(dfActual)

```

```python
dfTarget = spark.read.format("csv").option("header","true").load("Files/MtoMTarget.csv")
dfTarget = dfTarget.select(dfTarget.Country, dfTarget.Target.cast('int')) \
                    .groupBy("Country").sum("Target").withColumnRenamed("sum(Target)", "TargetTotal")
dfTarget.write.mode("overwrite").format("delta").saveAsTable("MtoMTargetSum")
display(dfTarget)
```

```python
# 另外一种创建表单的方法
from pyspark.sql import SparkSession
 
spark = SparkSession.builder.getOrCreate()
 
dfactual_data = [
    {'Country': 'France', 'ActualTotal': 4000},
    {'Country': 'Italy', 'ActualTotal': 16000},
    {'Country': 'England', 'ActualTotal': 23000}
]
 
dftarget_data = [
    {'Country': 'Germany', 'TargetTotal': 12000},
    {'Country': 'France', 'TargetTotal': 10500},
    {'Country': 'England', 'TargetTotal': 15000}
]
 
dfactual = spark.createDataFrame(dfactual_data)
dftarget = spark.createDataFrame(dftarget_data)
 
display(dfactual)
display(dftarget)
```

```python
dfbridge = dfActual.select("Country") \
                   .union(dfTarget.select("Country")) \
                   .distinct() \
                   .orderBy("Country")
display(dfbridge)
```

```python
dfjoin = dfbridge.join(dfActual, "Country", "left").join(dfTarget, "Country", "left")
display(dfjoin)
```
