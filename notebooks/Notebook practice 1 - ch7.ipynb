```python
df = spark.read.table("dimgeography")
display(df)
```

```python
df2 = df.select(df.City, df.StateProvinceName.alias("State"))
display(df2)
```

```python
#  where GeographyKey is equal to or greater than 200.
df3 = df.where(df.GeographyKey >= 200)
display(df3)
```

```python
df3 = df3.withColumn("IPLocation", df.IpAddressLocator.substr(1,3))
display(df3)
```

```python
display(df3.filter(df.CountryRegionCode.isin("FR", "GB")))
```

```python
%%sql
SELECT City, StateProvinceName AS State
FROM dimgeography;

SELECT *,
    substr(IpAddressLocator, 1, 3) AS IPLocation
FROM dimgeography
WHERE GeographyKey >= 200 AND CountryRegionCode IN ("FR", "GB");
```
