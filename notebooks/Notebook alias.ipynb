```python
df = spark.sql("SELECT * FROM mylakehouse.MtoMActual")
display(df)
```

```python
# dfreduced = df.select('Country', 'Actual') # not case sensitive
dfreduced = df.select(df.Country, df.Actual.alias('ActualSales')) # is case sensitive
# 当需要alias的时候使用下面的方式会更方便
# 'Actual'.alias是会报错
display(dfreduced)
```
